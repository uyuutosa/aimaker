```python
def load_setting(path):
    setting_dic = {}
    for path in lst:
        setting_dic.update(load_setting_rec(path, setting_dic=setting_dic))
    return setting_dic

def load_setting_rec(path, setting_dic, strip=''):
    splitted = path.strip(strip).split('/')
    if len(splitted) == 2:
        name = splitted[-1].strip('.json')
        #print(path)
        setting_dic[name] = json.load(open(path))
    else:
        name = splitted[1]
        try:
            setting_dic[name].update(load_setting_rec(path,  {}, strip=splitted[0] + '/'))
        except:
            setting_dic[name] = {}
            setting_dic[name].update(load_setting_rec(path,  {}, strip=splitted[0] + '/'))
    return setting_dic
    
```
